```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www.banki.ru/products/corporate/news/lenta/';
```

```python
r = requests.get(url);
print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'lxml');

dates = soup.find_all('div', class_ ='widget__date');
print(dates);
```
